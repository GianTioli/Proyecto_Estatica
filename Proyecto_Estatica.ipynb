```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Coordenadas de los puntos
A = np.array([0, 0, -60])
B = np.array([-36, -27, 10])
C = np.array([0, 32, 10])
D = np.array([40, -27, 10])
Peso = np.array([0, 0, -30])  # Aumentamos el tamaño del vector de peso

# Preparación de la figura y el eje
fig = plt.figure(figsize=(12, 10))
ax = fig.add_subplot(111, projection='3d')

# Dibuja las fuerzas como vectores desde A, manteniendo la proporción de las coordenadas
ax.quiver(A[0], A[1], A[2], B[0]-A[0], B[1]-A[1], B[2]-A[2], color='blue', label='Fuerza AB', arrow_length_ratio=0.1)
ax.quiver(A[0], A[1], A[2], C[0]-A[0], C[1]-A[1], C[2]-A[2], color='red', label='Fuerza AC', arrow_length_ratio=0.1)
ax.quiver(A[0], A[1], A[2], D[0]-A[0], D[1]-A[1], D[2]-A[2], color='green', label='Fuerza AD', arrow_length_ratio=0.1)
ax.quiver(A[0], A[1], A[2], Peso[0], Peso[1], Peso[2], color='black', label='Peso', arrow_length_ratio=0.1)

# Marcando las coordenadas de los puntos
for point, name in zip([A, B, C, D], ['A (0, -60, 0)', 'B (-36, 10, -27)', 'C (0, 10, 32)', 'D (40, 10, -27)']):
    ax.text(point[0], point[1], point[2] + 1.5, name, color='darkred', size=12)

# Configuración del gráfico
ax.set_xlim([-50, 50])
ax.set_ylim([-50, 50])
ax.set_zlim([-50, 50])
ax.set_xlabel('EJE X', fontsize=14, fontweight='bold')
ax.set_ylabel('EJE Z', fontsize=14, fontweight='bold')
ax.set_zlabel('EJE Y', fontsize=14, fontweight='bold')
ax.legend(loc='upper left', fontsize=12)

# Ajuste de la visualización
ax.view_init(elev=25., azim=100)

plt.title('Diagrama de cuerpo libre con fuerzas coordenadas', fontsize=16, fontweight='bold')
plt.show()



```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Datos proporcionados
delta_x = np.array([0.0092, 0.012, 0.014, 0.017, 0.02])  # Elongación Δx (m)
fuerza = np.array([0.18544, 0.26352, 0.3416, 0.41968, 0.49776])  # Fuerza F (N)

# Realizar ajuste lineal
slope, intercept, r_value, p_value, std_err = linregress(delta_x, fuerza)

# Crear valores para la línea de ajuste
x_fit = np.linspace(min(delta_x), max(delta_x), 100)
y_fit = slope * x_fit + intercept

# Crear gráfico
plt.figure(figsize=(8, 6))
plt.scatter(delta_x, fuerza, color='blue', label='Datos experimentales')
plt.plot(x_fit, y_fit, 'r-', label=f'Ajuste lineal: y = {slope:.6f}x + {intercept:.4f}')
plt.xlabel('Δx (m)')
plt.ylabel('Fuerza (N)')
plt.title('Relación entre Fuerza y Elongación con Ajuste Lineal')
plt.legend()
plt.grid(True)
plt.show()

# Imprimir la pendiente y el intercepto
print("Pendiente de la recta de ajuste:", slope)
print("Intercepto de la recta de ajuste:", intercept)
```
